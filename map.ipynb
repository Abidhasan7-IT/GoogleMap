```python
#first install pip folium
import folium
from IPython.display import display 

map_center = [23.758811, 90.399835]
 
mymap= folium.Map(location= map_center, zoom_start=12)
 
folium.Marker(
     [23.758811, 90.399835],
     popup= "Bangladesh",
     icon= folium.Icon(color="green",icon= "info-sign")
 ).add_to(mymap)
 

display(mymap)
```
